```python
import serial.tools.list_ports
import wmi

def list_com_ports():
    print("Listing available COM ports...")
    ports = serial.tools.list_ports.comports()
    if not ports:
        print("No COM ports found.")
    for port, desc, hwid in sorted(ports):
        print(f"{port}: {desc} [{hwid}]")

def list_usb_devices():
    print("Listing USB devices...")
    wmi_service = wmi.WMI()
    usb_devices = wmi_service.query("SELECT * FROM Win32_USBHub")

    if not usb_devices:
        print("No USB devices found.")
    for device in usb_devices:
        print(f"DeviceID: {device.DeviceID}")
        print(f"Description: {device.Description}")
        print(f"PNPDeviceID: {device.PNPDeviceID}")
        print(f"SystemName: {device.SystemName}")
        print()

if __name__ == "__main__":
    list_com_ports()
    list_usb_devices()
```

```python


# Replace 'COM3' with the appropriate port name found in the Device Manager
port = 'COM3'
baudrate = 9600  # Set the baud rate as per your device's specification

try:
    # List available COM ports
    print("Available COM ports:")
    list_com_ports()
    
    # Open the serial port
    ser = serial.Serial(port, baudrate)
    print(f"Connected to {port} at {baudrate} baud")

    # Continuously read data from the serial port
    while True:
        if ser.in_waiting > 0:
            data = ser.readline().decode('utf-8').rstrip()
            print(f"Received data: {data}")

except serial.SerialException as e:
    print(f"Error: {e}")

except KeyboardInterrupt:
    # Gracefully handle the user pressing Ctrl+C to exit the script
    print("Exiting...")

finally:
    # Close the serial port if it's open
    if ser.is_open:
        ser.close()
        print(f"Closed connection to {port}")
```

```python
import usb.core
import usb.util

# Vendor and Product IDs for the USB mouse
VENDOR_ID = 0x18F8
PRODUCT_ID = 0x1286 

# Find the USB device
device = usb.core.find(idVendor=VENDOR_ID, idProduct=PRODUCT_ID)

if device is None:
    raise ValueError("Device not found")

try:
    # Detach kernel driver if active (for Linux and macOS)
    if device.is_kernel_driver_active(0):
        device.detach_kernel_driver(0)

    # Set configuration
    device.set_configuration()

    # Claim interface
    usb.util.claim_interface(device, 0)

    # Read data from the mouse
    while True:
        # Read data from the endpoint (assuming endpoint 0)
        data = device.read(0x81, 8, timeout=100)
        
        # Print the raw data
        print("Raw data:", data)
        
        # Add your processing code here
        
finally:
    # Release the interface
    usb.util.release_interface(device, 0)

    # Reattach the kernel driver
    device.attach_kernel_driver(0)
```
